**import the python libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

**Read the dataset to python environment**

```python
diamonds =pd.read_csv('/content/diamonds.csv')
```

```python
diamonds.info()
```

```python
diamonds.head()
```

# **Histogram plot of various cuts**

```python
plt.hist(diamonds['cut'],bins=10,color='orange',edgecolor='k')
plt.xlabel('Cuts',fontsize=15)
plt.ylabel('Counts',fontsize=15)
plt.title('Histogram of various cuts')
plt.show()
```

# **Bar plot with cut,price and color**

```python
sns.barplot(x='cut',y='price',hue='color',data=diamonds)
plt.title('Cut,Price and color',fontsize=15)
plt.show()
```

**Bar plot of cut,price and clarity**

```python
sns.barplot(x='cut',y='price',hue='clarity',data=diamonds)
plt.title('Cut , Price and Clarity')
plt.legend()
plt.grid(True)
plt.show()
```

## **KDE Histplot of Price**

```python
sns.histplot(diamonds['price'],kde=True)
```

# **Jointplot  of  price and carat**

```python
sns.jointplot(x='price',y='carat',data=diamonds)
```

# ***`Correlation matrix`***

```python
plt.title('Correlation matrix')
sns.heatmap(diamonds.corr(numeric_only=True),annot=True)
```
